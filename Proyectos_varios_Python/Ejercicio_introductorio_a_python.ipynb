```python
def mayus_to_minus(string):
    # Convierte una cadena a minúsculas
    string = string.lower()
    return string

def minus_to_mayus(string):
    # Convierte una cadena a mayúsculas
    string = string.upper()
    return string
```

```python
# prompt: crea una función que asigne de forma aleatoria a los estudiantes en grupos de máximo 6 personas, con un máximo de 27 grupos y que el grupo quede en la variable Grupo

import pandas as pd
import numpy as np
def asignar_grupos_aleatorios(df):
    """
    Asigna aleatoriamente a los estudiantes en grupos de máximo 6 personas,
    con un máximo de 27 grupos.

    Args:
        df (pd.DataFrame): DataFrame con una columna 'Grupo'.

    Returns:
        pd.DataFrame: DataFrame con la columna 'Grupo' actualizada.
    """
    # Número de estudiantes
    num_estudiantes = len(df)

    # Calcular el número mínimo de grupos necesarios
    min_grupos = m.ceil(num_estudiantes / 6)

    # Asegurarse de no exceder el máximo de 27 grupos
    num_grupos = min(min_grupos, 27)

    # Crear una lista con los identificadores de grupo
    grupos = [f'{i+1}' for i in range(num_grupos)]

    # Asignar grupos aleatoriamente a los estudiantes
    df['Grupo'] = np.random.choice(grupos, size=num_estudiantes, replace=True)

    # Ajustar el tamaño de los grupos para que no superen 6
    for grupo in grupos:
        estudiantes_en_grupo = df[df['Grupo'] == grupo]
        if len(estudiantes_en_grupo) > 6:
            # Seleccionar aleatoriamente estudiantes para mover a otros grupos
            estudiantes_a_mover = estudiantes_en_grupo.sample(n=len(estudiantes_en_grupo) - 6, replace=False)
            # Asignarles grupos aleatorios nuevamente
            df.loc[estudiantes_a_mover.index, 'Grupo'] = np.random.choice(grupos, size=len(estudiantes_a_mover), replace=True)

    return df

df_estudiantes = asignar_grupos_aleatorios(df_estudiantes)
df_estudiantes
```

```python
# prompt: equilibra los grupos asignando personas de los grupos con 7 y 8 personas a grupos con menos de tal forma que no queden grupos con más de 6 estudiantes

import numpy as np
# Identificar grupos con más de 6 estudiantes
grupos_grandes = df_estudiantes['Grupo'].value_counts()
grupos_grandes = grupos_grandes[grupos_grandes > 6].index.tolist()

# Identificar grupos con menos de 6 estudiantes
grupos_pequenos = df_estudiantes['Grupo'].value_counts()
grupos_pequenos = grupos_pequenos[grupos_pequenos < 6].index.tolist()

# Iterar sobre los grupos grandes y mover estudiantes a grupos pequeños
for grupo_grande in grupos_grandes:
    estudiantes_en_grupo_grande = df_estudiantes[df_estudiantes['Grupo'] == grupo_grande]
    num_estudiantes_a_mover = len(estudiantes_en_grupo_grande) - 6

    # Mover estudiantes si hay grupos pequeños disponibles
    if len(grupos_pequenos) > 0:
        # Seleccionar aleatoriamente estudiantes para mover
        estudiantes_a_mover = estudiantes_en_grupo_grande.sample(n=num_estudiantes_a_mover, replace=False)

        # Asignarles grupos pequeños aleatoriamente
        grupos_destino = np.random.choice(grupos_pequenos, size=len(estudiantes_a_mover), replace=True)
        df_estudiantes.loc[estudiantes_a_mover.index, 'Grupo'] = grupos_destino

        # Actualizar la lista de grupos pequeños (puede que algunos se llenen)
        grupos_pequenos = df_estudiantes['Grupo'].value_counts()
        grupos_pequenos = grupos_pequenos[grupos_pequenos < 6].index.tolist()

# Verificar el tamaño de los grupos después del equilibrio
print("Tamaño de los grupos después del equilibrio:")
print(df_estudiantes['Grupo'].value_counts())
```

```python
# prompt: muestrame los grupos con personas que tengan compartan al menos un nombre

# Obtener los nombres de cada estudiante
nombres_estudiantes = df_estudiantes['Nombres'].tolist()

# Crear un conjunto para almacenar nombres ya vistos
nombres_vistos = set()

# Crear un diccionario para agrupar estudiantes por nombre compartido
grupos_nombres_compartidos = {}

# Iterar sobre cada estudiante y su nombre
for index, row in df_estudiantes.iterrows():
    nombre_completo = row['Nombres']
    grupo_estudiante = row['Grupo']

    # Dividir el nombre en nombres individuales
    nombres_individuales = nombre_completo.split()

    # Verificar si algún nombre individual ya fue visto en otro grupo
    nombres_compartidos_encontrados = set()
    for nombre in nombres_individuales:
        if nombre in nombres_vistos:
            nombres_compartidos_encontrados.add(nombre)
        else:
            nombres_vistos.add(nombre)

    # Si se encontraron nombres compartidos, agrupar al estudiante
    if nombres_compartidos_encontrados:
        for nombre_compartido in nombres_compartidos_encontrados:
            if nombre_compartido not in grupos_nombres_compartidos:
                grupos_nombres_compartidos[nombre_compartido] = set()
            grupos_nombres_compartidos[nombre_compartido].add(grupo_estudiante)

# Imprimir los grupos con personas que comparten al menos un nombre
print("Grupos con personas que comparten al menos un nombre:")
for nombre, grupos in grupos_nombres_compartidos.items():
    if len(grupos) > 1: # Solo mostrar si hay más de un grupo compartiendo el nombre
        print(f"Nombre compartido '{nombre}': Grupos {list(grupos)}")

# Opcional: Mostrar los estudiantes dentro de esos grupos
for nombre, grupos in grupos_nombres_compartidos.items():
     if len(grupos) > 1:
        print(f"\nEstudiantes con el nombre compartido '{nombre}':")
        for grupo in grupos:
            estudiantes_en_grupo = df_estudiantes[(df_estudiantes['Grupo'] == grupo) &
                                                  (df_estudiantes['Nombres'].str.contains(nombre, case=False))]
            print(f"  Grupo {grupo}:")
            for idx, estudiante_row in estudiantes_en_grupo.iterrows():
                 print(f"    - {estudiante_row['Nombres']}")
```
